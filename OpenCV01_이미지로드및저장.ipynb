## OpenCV 개요 및 환경설정

### 개요

- <b>OpenCV </b>
  - Gray Bradsky에 의해 1996년 인텔에서 시작된 프로젝트
  - 컴퓨터 비전, 머신러닝과 관련된 다양한 알고리즘을 지원
  - C++, Python, Java 등의 언어를 지원
  - CUDA(Compute Unified Device Architecture), OpenCL (Open Computing Language)에 기반한 인터페이스를 지원
  <br><br>
- <b>OpenCV + Python</b> 
  - OpenCV의 파이썬 API로 C/C++로 된 OpenCV 라이브러리들을 파이썬 래퍼로 감싼 후에 파이썬 모듈을 추가시킨 것
  - 속도문제를 해결하기 위해 속도가 문제되는 코드는 C/C++로 코딩하고 파이썬에서 불러 사용할 수 있는 파이썬 래퍼를 제공
  - OpenCV 배열은 Numpy 배열로 변환되어 내부 처리를 수행 (Numpy의 모든 연산 가능)
  - Numpy, SciPy, Matplotlib 라이브러리와도 호환

## 이미지를 읽고 출력하기

  - <font color=red>cv2.imread(파일명, 이미지 형식)</font> : 이미지 파일을 읽기 위한 객체를 리턴 
    - 이미지형식 : 칼라 (cv2.IMREAD_COLOR), 흑백(cv2.IMREAD_GRAYSCALE) 등
  - <font color=red>cv2.imshow(제목, 이미지 객체)</font>  : 이미지 출력
  - <font color=red>cv2.waitKey(0)</font>  : 키보드 입력을 기다리는 시간 설정 (0 : 계속 기다림)
    - 1/1000초 단위
  - <font color=red>cv2.destroyAllWindows()</font>  : 생성한 윈도우를 모두 닫는다

<img src="../OpenCV_data/lecture_image/01_lenna.png" width=30%>

### window를 이용하여 이미지 출력하기

```python
# 윈도우 창에 이미지 띄우기
# 윈도우 창 닫는 법 모름 ㅠ
import cv2
img = cv2.imread("../OpenCV_data/image/suzy.jfif", cv2.IMREAD_COLOR)  # 이미지 읽어서 객체에 담기
cv2.namedWindow('suzy', flags=None)  # (윈도우이름, flags설정)이미지를 표시할 수 있는 윈도우 생성
cv2.resizeWindow('suzy',200,400)  # (윈도우이름,너비,높이)윈도우 사이즈 조정
cv2.imshow('suzy',img)
cv2.waitKey(0) 
cv2.destroyWindow('suzy') 
```

### matplotlib를 이용하여 이미지 출력하기

```python
!pip install matplotlib
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("../OpenCV_data/image/suzy.jfif", cv2.IMREAD_COLOR)
plt.imshow(img)
```

- OpenCV는 색상공간을 BGR(Blue Green Red)을 사용
- Python 이미지들은 RGB(Red Green Blue)을 사용하므로 변환 필요
- 따라서, OpenCV로 읽은 이미지를 Matplotlib으로 출력하려면 색상공간을 변환해야 함

<img src="../OpenCV_data/lecture_image/01_color_space.png" width=40%>

- <font color=red>cv2. cvtColor(img, 색상변환옵션)</font>
  - 색상변환옵션 : cv2.COLOR_BGR2RGB, cv2.COLOR_RGB2GRAY, cv2.COLOR_RGB2HSV 등

```python
# 색상 공간 변경
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# 축 눈금을 제거
plt.xticks([]); plt.yticks([])
# 그래프의 제목
plt.title("suzy")
# 이미지 출력
plt.imshow(img2)
plt.show()
```

#### 직접 Red 채널과 Blue 채널을 바꾸기

- 채널(chnnel) : 색상공간을 구성하는 색상을 의미, 칼라이미지는 Red 이미지, Green 이미지, Blue 이미지의 3개의 이미지로 구성

-  <font color=red>b, g, r = cv2.split(img)</font> : img파일을 b,g,r로 분리
-  <font color=red>cv2.merge([r,g,b])</font> : b, r을 바꿔서 Merge

```python
b, g, r = cv2.split(img)  # 이미지 채널을 나눠서 
img3 = cv2.merge([r,g,b])  # 재조합
plt.xticks([]); plt.yticks([])
plt.imshow(img2)
plt.show()
```

#### 각 채널 이미지 확인

```python
# 여러개의 그래프를 테이블 형태로 배치하는 방법 (행의 개수, 열의 개수, 번호)
plt.subplot(1, 3, 1)
plt.xticks([]); plt.yticks([])
plt.title("Red")
plt.imshow(r, cmap="gray") # cmap="gray" : 흑백이미지로 출력

plt.subplot(1, 3, 2)
plt.xticks([]); plt.yticks([])
plt.title("Green")
plt.imshow(g, cmap="gray")

plt.subplot(1, 3, 3)
plt.xticks([]); plt.yticks([])
plt.title("Blue")
plt.imshow(b, cmap="gray")
plt.show()

# 채널에 흰색 (255) 쪽 성분이 많을수록 해당 색상이 많이 분포되어 있다는 것을 의미
```

### Widget을 이용해서 이미지 출력하기
- widget.Image(format=' ', width= , height= ) : 이미지를 출력하는 위젯
- format=".jepg" : 출력할 이미지 타입
- width, height : 출력할 이미지의 크기

```python
!pip install ipywidgets
```

```python
# ipywidgets : UI 라이브러리
import ipywidgets as widget
# 이미지나 소리 등을 출력하는 함수
from IPython.display import display

img_widget = widget.Image(format=".jpepg", width=300, height=300)
display(img_widget)
_, en_img = cv2.imencode(".jpeg", img)  # OpenCV에 입력된 이미지는 Numpy array라서 jpeg로 변환해주어야 한다

img_widget.value = en_img.tobytes()
```
